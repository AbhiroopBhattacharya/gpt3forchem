```python
import pandas as pd 
import EFGs

from gpt3forchem.data import get_esol_data, get_freesolv_data, get_lipophilicity_data, get_photoswitch_data
from rdkit import Chem

import seaborn as sns
import matplotlib.pyplot as plt

import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)

plt.style.use(['science', 'nature'])
from gpt3forchem.helpers import get_num_token_dist, get_representation_length_dist, get_token_counts

def get_num_token_and_len_dist_and_token_counts(representations): 
    return get_num_token_dist(representations), get_representation_length_dist(representations), get_token_counts(representations)
```

```python
esol_data = get_esol_data()
freesolv_data = get_freesolv_data()
lipophilicity_data = get_lipophilicity_data()
```

```python
photoswitch_data = get_photoswitch_data()
```

```python
efgs_fragments = [EFGs.mol2frag(Chem.MolFromSmiles(mol)) for mol in esol_data['smiles']]
```

```python
freesolv_fragments = [EFGs.mol2frag(Chem.MolFromSmiles(mol)) for mol in freesolv_data['smiles']]
```

```python
lipophilicity_fragments = [EFGs.mol2frag(Chem.MolFromSmiles(mol)) for mol in lipophilicity_data['smiles']]
```

```python
photoswitch_fragments = [EFGs.mol2frag(Chem.MolFromSmiles(mol)) for mol in photoswitch_data['SMILES']]
```

```python
efgs_number_of_fragments = [len(frag[0]) for frag in efgs_fragments]
freesolv_number_of_fragments = [len(frag[0]) for frag in freesolv_fragments]
lipophilicity_number_of_fragments = [len(frag[0]) for frag in lipophilicity_fragments]
photoswitch_number_of_fragments = [len(frag[0]) for frag in photoswitch_fragments]
```

```python
sns.distplot(efgs_number_of_fragments,  label='ESOL')
sns.distplot(freesolv_number_of_fragments,label='FreeSolv')
sns.distplot(lipophilicity_number_of_fragments,  label='Lipophilicity')
sns.distplot(photoswitch_number_of_fragments,  label='Photoswitch')
plt.legend()
```

```python
smiles_num_tokens_esol, smiles_len_esol, smiles_token_counts_esol = get_num_token_and_len_dist_and_token_counts(esol_data['smiles'])

selfies_num_tokens_esol, selfies_len_esol, selfies_token_counts_esol = get_num_token_and_len_dist_and_token_counts(esol_data['selfies'])

name_num_tokens_esol, name_len_esol, name_token_counts_esol = get_num_token_and_len_dist_and_token_counts(esol_data['iupac_name'])
```

```python

smiles_num_tokens_freesolv, smiles_len_freesolv, smiles_token_counts_freesolv = get_num_token_and_len_dist_and_token_counts(freesolv_data['smiles'])

selfies_num_tokens_freesolv, selfies_len_freesolv, selfies_token_counts_freesolv = get_num_token_and_len_dist_and_token_counts(freesolv_data['selfies'])

name_num_tokens_freesolv, name_len_freesolv, name_token_counts_freesolv = get_num_token_and_len_dist_and_token_counts(freesolv_data['iupac_name'])
```

```python

smiles_num_tokens_lipophilicity, smiles_len_lipophilicity, smiles_token_counts_lipophilicity = get_num_token_and_len_dist_and_token_counts(lipophilicity_data['smiles'])

selfies_num_tokens_lipophilicity, selfies_len_lipophilicity, selfies_token_counts_lipophilicity = get_num_token_and_len_dist_and_token_counts(lipophilicity_data['selfies'])

name_num_tokens_lipophilicity, name_len_lipophilicity, name_token_counts_lipophilicity = get_num_token_and_len_dist_and_token_counts(lipophilicity_data['iupac_name'])
```

```python



smiles_num_tokens_photoswitch, smiles_len_photoswitch, smiles_token_counts_photoswitch = get_num_token_and_len_dist_and_token_counts(photoswitch_data['SMILES'])

selfies_num_tokens_photoswitch, selfies_len_photoswitch, selfies_token_counts_photoswitch = get_num_token_and_len_dist_and_token_counts(photoswitch_data['selfies'])

name_num_tokens_photoswitch, name_len_photoswitch, name_token_counts_photoswitch = get_num_token_and_len_dist_and_token_counts(photoswitch_data.dropna(subset='name')['name'])
```

```python
fig, ax = plt.subplots(1, 4, sharey=True, figsize=(7,3))


sns.distplot(smiles_num_tokens_esol, label='SMILES', ax=ax[0])
sns.distplot(selfies_num_tokens_esol, label='SELFIES', ax=ax[0])
sns.distplot(name_num_tokens_esol, label='name', ax=ax[0])

sns.distplot(smiles_num_tokens_freesolv, label='SMILES', ax=ax[1])
sns.distplot(selfies_num_tokens_freesolv, label='SELFIES', ax=ax[1])
sns.distplot(name_num_tokens_freesolv, label='name', ax=ax[1])

sns.distplot(smiles_num_tokens_lipophilicity, label='SMILES', ax=ax[2])
sns.distplot(selfies_num_tokens_lipophilicity, label='SELFIES', ax=ax[2])
sns.distplot(name_num_tokens_lipophilicity, label='name', ax=ax[2])

sns.distplot(smiles_num_tokens_photoswitch, label='SMILES', ax=ax[3])
sns.distplot(selfies_num_tokens_photoswitch, label='SELFIES', ax=ax[3])
sns.distplot(name_num_tokens_photoswitch, label='name', ax=ax[3])

ax[0].set_ylim(0, 0.08)
ax[0].legend()
ax[0].set_title('ESOL')
ax[1].set_title('FreeSolv')
ax[2].set_title('Lipophilicity')
ax[3].set_title('Photoswitch')

for i in range(4):
    ax[i].set_xlabel('number of tokens')

fig.tight_layout()
fig.savefig('num_tokens_dist.pdf', bbb_inches='tight')
```

```python
fig, ax = plt.subplots(1, 4, sharey=True, figsize=(7,3))

# now plot the length distribution
sns.distplot(smiles_len_esol, label='SMILES', ax=ax[0])
sns.distplot(selfies_len_esol, label='SELFIES', ax=ax[0])
sns.distplot(name_len_esol, label='name', ax=ax[0])

sns.distplot(smiles_len_freesolv, label='SMILES', ax=ax[1])
sns.distplot(selfies_len_freesolv, label='SELFIES', ax=ax[1])
sns.distplot(name_len_freesolv, label='name', ax=ax[1])

sns.distplot(smiles_len_lipophilicity, label='SMILES', ax=ax[2])
sns.distplot(selfies_len_lipophilicity, label='SELFIES', ax=ax[2])
sns.distplot(name_len_lipophilicity, label='name', ax=ax[2])


sns.distplot(smiles_len_photoswitch, label='SMILES', ax=ax[3])
sns.distplot(selfies_len_photoswitch, label='SELFIES', ax=ax[3])
sns.distplot(name_len_photoswitch, label='name', ax=ax[3])

ax[0].set_ylim(0, 0.08)
ax[0].legend()
ax[0].set_title('ESOL')
ax[1].set_title('FreeSolv')
ax[2].set_title('Lipophilicity')
ax[3].set_title('Photoswitch')

for i in range(4):
    ax[i].set_xlabel('length of sequence')

fig.tight_layout()
fig.savefig('length_dist.pdf', bbb_inches='tight')

```

```python
fig, ax = plt.subplots(1, 4, sharey=True)

# now plot the count distribution

sns.distplot(smiles_token_counts_esol, label='SMILES', ax=ax[0])
sns.distplot(selfies_token_counts_esol, label='SELFIES', ax=ax[0])
sns.distplot(name_token_counts_esol, label='name', ax=ax[0])

sns.distplot(smiles_token_counts_freesolv, label='SMILES', ax=ax[1])
sns.distplot(selfies_token_counts_freesolv, label='SELFIES', ax=ax[1])
sns.distplot(name_token_counts_freesolv, label='name', ax=ax[1])

sns.distplot(smiles_token_counts_lipophilicity, label='SMILES', ax=ax[2])
sns.distplot(selfies_token_counts_lipophilicity, label='SELFIES', ax=ax[2])
sns.distplot(name_token_counts_lipophilicity, label='name', ax=ax[2])

sns.distplot(smiles_token_counts_photoswitch, label='SMILES', ax=ax[3])
sns.distplot(selfies_token_counts_photoswitch, label='SELFIES', ax=ax[3])
sns.distplot(name_token_counts_photoswitch, label='name', ax=ax[3])

ax[0].set_ylim(0, 0.08)
ax[0].legend()
ax[0].set_title('ESOL')
ax[1].set_title('FreeSolv')
ax[2].set_title('Lipophilicity')
ax[3].set_title('Photoswitch')
```
